# Procesamiento de datos

Este script tiene como objetivo procesar los datos extraídos de las diferentes ligas de futbol del mundo realizados en el paso anterior mediante *Scraping, con el fin de prepararlos para implementación de modelo de Machine Learning que realizara predicciones como por ejemplo tiros de esquina que se realizaran después del minuto 75 entre otros.

Se realizarán las siguientes acciones:

•	Unión de diferentes archivos csv en uno solo.

•	Eliminar duplicados en caso de existir.

•	Extraer los totales.

•	Agrupar datos para calcular puntos llevados por cada equipo de acuerdo a la fecha.

•	Merged entre dataframe para signar puntos a Data original.

•	Creacion de archivos csv

## Unión de diferentes archivos csv en uno solo

dado que la información de cada liga por temporadas viene en diferentes archivos es necesario unirla en uno solo

```python
# Ruta de la carpeta que contiene los archivos CSV
carpeta = 'C:\\Users\\sandr\\OneDrive\\Escritorio\\phyton_2023\\10_futbol\\archivos_futbol\\unir_data_frame'

# Lista para almacenar los DataFrames de los archivos CSV
dfs = []

# Iterar sobre todos los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):
        # Leer el archivo CSV y almacenarlo en un DataFrame
        ruta_archivo = os.path.join(carpeta, archivo)
        df = pd.read_csv(ruta_archivo)
        dfs.append(df)
        
    print('finalizo el cargue')
df_final = pd.concat(dfs, ignore_index=True)
```

## Agrupar datos para calcular puntos llevados por cada equipo de acuerdo a la fecha

```python
resumen_por_grupo = df_invertido.groupby(['equipo','Temporadas', 'Fecha'], sort=False)['Goals'].sum().reset_index()
```

```python
resumen_por_grupo
```

funcion para sacar puntos por partido

```python
def calcular_puntos(fila1, fila2):
    if fila1['Goals'] > fila2['Goals']:
        return 3
    elif fila1['Goals'] < fila2['Goals']:
        return 0
    else:
        return 1
```

```python
puntos = []
for i in range(0, len(resumen_por_grupo), 2):
    if i + 1 < len(resumen_por_grupo):  # Evitar índices fuera de rango
        puntos.append(calcular_puntos(resumen_por_grupo.iloc[i], resumen_por_grupo.iloc[i + 1]))
        puntos.append(calcular_puntos(resumen_por_grupo.iloc[i + 1],resumen_por_grupo.iloc[i]))
    else:  # Si hay un número impar de filas, la última no se compara
        puntos.append(0)

#Asignar los puntos a cada fila del DataFrame
resumen_por_grupo['Puntos'] = puntos
```

```python
resumen_por_grupo
```

cogigo para calcular puntos acumulados de acuerdo a temporada

```python
resumen_por_grupo['Puntos_Acum'] = resumen_por_grupo.groupby(['Temporadas', 'equipo'])['Puntos'].cumsum()
```

```python
resumen_por_grupo
```

codigo para calcular goles recibidos

```python
GOL_MARC = []
for i in range(0, len(resumen_por_grupo), 2):
    if i + 1 < len(resumen_por_grupo):  # Evitar índices fuera de rango
        GOL_MARC.append(resumen_por_grupo.iloc[i + 1]['Goals'])
        GOL_MARC.append(resumen_por_grupo.iloc[i]['Goals'])
    else:  # Si hay un número impar de filas, la última no se compara
        GOL_MARC.append(0)

#Asignar los puntos a cada fila del DataFrame
resumen_por_grupo['Goals_recived'] = GOL_MARC
```

```python
resumen_por_grupo
```

calculo de gol diferencia

```python
resumen_por_grupo['Gol_Diferencia'] = resumen_por_grupo['Goals'] -  resumen_por_grupo['Goals_recived']
```

```python
resumen_por_grupo
```

sumatoria para calcular gol diferencia por temporada

```python
resumen_por_grupo['Gol_Dif_acum'] = resumen_por_grupo.groupby(['Temporadas', 'equipo'])['Gol_Diferencia'].cumsum()
```

```python
resumen_por_grupo
```

creación de listas para realizar ordenamientos

```python
Temporadas = resumen_por_grupo['Temporadas'].unique()
```

```python
Fechas = resumen_por_grupo['Fecha'].unique()
```

ciclo para ordenar por temporada y fecha

```python

df_posicion = []

# Iterar sobre todos los archivos en la carpeta
for Temporada in Temporadas :
    for Fecha in Fechas:
        df_provi = resumen_por_grupo.loc[(resumen_por_grupo['Temporadas'] == Temporada) & (resumen_por_grupo['Fecha'] == Fecha)]
        df_provi = df_provi.sort_values(by=['Puntos_Acum', 'Gol_Dif_acum'], ascending=[False, False])
        df_provi = df_provi.reset_index(drop=True)
        df_provi['Posición'] = df_provi.reset_index().index + 1   
        df_posicion.append(df_provi)   
        print(f'finalizo el procesamiento temporada: {Temporada} y fecha {Fecha}')
df_posicion= pd.concat(df_posicion, ignore_index=True)
```

## Merged entre dataframe para signar puntos a Data original

se concatenan columnas a dataframe que contiene las posiciones para crear una sola y poder realizar cruce

```python
df_posicion['col_comun'] = df_posicion['equipo'] + df_posicion['Temporadas'] + df_posicion['Fecha']
```

```python
df_posicion
```

se hace lo mismo para el DataFrame inicial que contienen información completa sin duplicados

```python
df_invertido['col_comun'] = df_invertido['equipo'] + df_invertido['Temporadas'] + df_invertido['Fecha']
```

```python
df_invertido
```

se realiza un Merged para dejar un Dataframe final que contiene toda la información inicial sin duplicados agregando la posición y gol diferencia de cada equipo por partido

```python
df_posicion.loc[(df_posicion['Temporadas'] == '2022/2023') & (df_posicion['Fecha'] == 'Matchday 38')]
```

```python
df_final_puntos = pd.merge(df_invertido, df_posicion[['col_comun', 'Posición', 'Gol_Dif_acum']], on='col_comun', how='left')
```
